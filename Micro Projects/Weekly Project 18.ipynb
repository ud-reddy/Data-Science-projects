```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('product.csv')
```

```python
df
```

```python
def dataframe(df, x1, x2):
    
    x = list(df[x1])
    y = list(df[x2])
    
    data = []
    for i in range(len(x)):
        data.append([x[i],y[i]])
    
    df_ = pd.DataFrame(data,columns=[x1,x2])
    
    return df_
```

## 1) Name and Order Analysis

```python
df_ = dataframe(df, 'Name', 'Orders')

x = df_.sort_values(by='Orders', ascending = False)['Name']
y = df_.sort_values(by='Orders', ascending = False)['Orders']

fig, axs = plt.subplots(figsize=(10,6))
axs.barh(x,y)
plt.xlabel('Products')
plt.ylabel('Number Of Orders')
plt.title('Number of Orders per Product')
plt.show()
```

#### Conclusion 1 : By the above graph we can conclude that "USB-C Charging Cable" was ordered the most followed by "Lightning Charging Cable" and "AAA Batteries(4-pack)" , while "LG Dryer" is the least ordered Product

## 2) Name and Max Quantity Per User(Column) Analysis

```python
df_ = dataframe(df, 'Name', 'Max_quantity_per_user')

x = df_.sort_values(by='Max_quantity_per_user', ascending = False)['Name']
y = df_.sort_values(by='Max_quantity_per_user', ascending = False)['Max_quantity_per_user']

fig, axs = plt.subplots(figsize=(10,6))
axs.barh(x,y)
plt.xlabel('Products')
plt.ylabel('Max Quantity Per User')
plt.title(' Maximum number of Quantity per user')
plt.show()
```

#### Conclusion 2: From the above graph we can conclude that 'AAA Batteries (4-pack)' has the maximum quantity per user while 'LG Dryer' has the least.

## 3) Name and Amount Analysis

```python
df_ = dataframe(df, 'Name', 'Amount')

x = df_.sort_values(by='Amount', ascending = False)['Name']
y = df_.sort_values(by='Amount', ascending = False)['Amount']

fig, axs = plt.subplots(figsize=(10,6))
axs.barh(x,y)
plt.xlabel('Products')
plt.ylabel('Amount')
plt.title('Total Amount of Each Product')
plt.show()
```

#### Conclusion 3: From the above graph we can conclude that 'Macbook Pro Laptop' has collect the most total amount followed by 'iphone' and 'ThinkPad Laptop', while 'AAA Batteries (4-pack)' has the least total amount collected after sales.

```python
# The Average Total Amount Per each order placed is made into seperate column called AvgAmountPerOrder

df['AvgAmountPerOrder'] = round(df['Amount']/df['Orders'],2)
df.head()
```

### 4.2) Adding Average Amount per each product sold column

```python
# The Average Total Amount per each Product (Quantity) sold is made into seperated column called AvgAmountPerProduct

df['AvgAmountPerProduct'] = round(df['Amount']/df['Quantity'],2)
df.head()
```

```python
# The Maximum Amount collect by each product is made into a seperate column called MaxAmountPerProduct

df['MaxAmountPerProduct'] = round(df['Price']*df['Max_quantity_per_user'],2)
df.head()
```
