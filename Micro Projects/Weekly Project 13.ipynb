```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("country_vaccinations.csv")
```

```python
data = df.values
```

# Multi Column Analysis

### Replacing the null values with zero

```python
df = df.fillna(0)
```

```python
df
```

## Total vaccinations in india in may 2021 alone

```python
print("-"*40)
print("The total vaccinations in India in May 2021 : ")
print('-'*40)
l=[]
for i in data:
    if(i[0]=='India' and i[15]=='2021' and i[17]=='May'):
        l.append([i[0],i[3]])
pdf=pd.DataFrame(l,columns=['country','total_vaccinations'])
print(pdf.sort_values(by='total_vaccinations',ascending=False).values[0][1])
print('-'*40)
```

## People vaccinated in Albania in 2021-01-12

```python
print('-'*40)
print('People vaccinated in Albania on 2021-01-12 : ')
print('-'*30)
for i in data:
    if(i[0]=='Albania' and i[2]=='2021-01-12'):
        print(i[4])
print('-'*40)
```

### Finding out the unique countries in the dataset 

```python
country = set(list(df['country']))
```

```python
print('-'*50)
print("The total number of unique countries are : ",len(country))
print('-'*50)
print("And they are : \n",country)
print('-'*50)
```

### Adding the year, month(jan -dec) and day columns to the dataset

```python
year=[]
lst = list(df['date'])
for i in range(len(lst)):
    year.append(lst[i].split('-')[0])
df['Year']=year

day = []
for i in range(len(lst)):
    day.append(lst[i].split('-')[2])
df['day'] = day

months = ['January','February',
         'March','April',
         'May','June',
         'July','August',
         'September','October',
         'November','December']
lst = list(df['date'])
mon=[]
for i in range(len(lst)):
    mon.append(months[int(lst[i].split('-')[1])-1])
df['Month']=mon
```
