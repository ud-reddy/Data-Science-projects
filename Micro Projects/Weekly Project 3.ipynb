## Importing the required modules

```python
import pandas as pd
```

## Loading the dataset 

```python
df = pd.read_csv("country_wise_latest.csv")
df.head()
```

## 1) The average cases in every country 

```python
print('The average cases per each country : ')
print("-"*30)
for i in df.values:
    print(i[0],' : ',round((i[1]+i[11])/2,0))
print("-"*30)
```

## 2) Total deaths as per the dataset

```python
print('The total number of deaths as per the dataset are : ')
print('-'*40)
print(df['Deaths'].sum())
print('-'*40)
```

## 3) Total number confirmed cases

```python
print('The total number of confirmed cases are : ')
print('-'*40)
print(df['Confirmed'].sum())
print('-'*40)
```
