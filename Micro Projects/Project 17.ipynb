```python
gender = []
for i in df['Gender']:
    if i=='Boy':
        gender.append(1)
    if i=='Girl':
        gender.append(0)
df['Gender']=gender

#boy = 1
#girl = 0
```

```python
age = []
for i in df['Age']:
    age.append(int(int(i.split('-')[0])+int(i.split('-')[1]))/2)
df['Age']=age    
```

```python
# university = 0, college  = 1, school = 2
edlevel = []
for i in df['Education Level']:
    if i == 'University':
        edlevel.append(0)
    elif i == 'College':
        edlevel.append(1)
    elif i == 'School':
        edlevel.append(2)
df['Education Level']=edlevel
```

```python
df['Institution Type'].unique()
# Non Government = 0, Government = 1 
itype = []
for i in df['Institution Type']:
    if i=='Non Government':
        itype.append(0)
    elif i=='Government':
        itype.append(1)
df['Institution Type'] = itype
```

```python
df['IT Student'].unique()

# yes = 1, no = 0

itstd = []
for i in df['IT Student']:
    if i=='Yes':
        itstd.append(1)
    elif i=='No':
        itstd.append(0)
df['IT Student'] = itstd
```

```python
df['Location'].unique()

# yes = 1, no = 0 

loc = []
for i in df['Location']:
    if i=='Yes':
        loc.append(1)
    elif i=='No':
        loc.append(0)
df['Location'] = loc
```

```python
df['Load-shedding'].unique()

#low = 0, high = 1

ls = []
for i in df['Load-shedding']:
    if i=='Low':
        ls.append(0)
    elif i=='High':
        ls.append(1)
df['Load-shedding'] = ls
```

```python
df['Financial Condition'].unique()

#poor = 0, mid =1, rich = 2 

fc = []
for i in df['Financial Condition']:
    if i=='Poor':
        fc.append(0)
    elif i=='Mid':
        fc.append(1)
    elif i=='Rich':
        fc.append(2)
df['Financial Condition'] = fc
```

```python
df['Internet Type'].unique()

# wifi = 0 , mobile data = 1

itype = []
for i in df['Internet Type']:
    if i=='Wifi':
        itype.append(0)
    elif i=='Mobile Data':
        itype.append(1)
df['Internet Type'] = itype
```

```python
df['Network Type'].unique()

# 2G = 0, 3G = 1 ,2G = 3

ntype = []
for i in df['Network Type']:
    if i=='2G':
        ntype.append(0)
    elif i=='3G':
        ntype.append(1)
    elif i=='4G':
        ntype.append(2)
df['Network Type'] = ntype
```
