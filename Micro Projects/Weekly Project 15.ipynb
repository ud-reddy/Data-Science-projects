```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("country_vaccinations.csv")
```

```python
months = ['January','February',
         'March','April',
         'May','June',
         'July','August',
         'September','October',
         'November','December']
lst = list(df['date'])
mon=[]
for i in range(len(lst)):
    mon.append(months[int(lst[i].split('-')[1])-1])
df['Month']=mon

year=[]
lst = list(df['date'])
for i in range(len(lst)):
    year.append(lst[i].split('-')[0])
df['Year']=year

day = []
for i in range(len(lst)):
    day.append(lst[i].split('-')[2])
df['day'] = day
df
```

```python
data = df.values
```

```python
l=[]
for i in data:
    if(i[16]=='2021'):
        l.append(i)
pdf=pd.DataFrame(l,columns=list(df.columns))
```

```python
month_end=[31,28,31,30,31,30,31,31,30,31,30,31]
```

```python
d={}
n=0
for i in months:
    l=[]
    for j in pdf.values:
        if(i==j[15]):
            if(j[17]==str(month_end[months.index(i)])):
                l.append(j[3])
    d[i]=sum(l)
print("-"*65)
print("The Monthly number of total vaccination from Jan to Dec in 2021 : ")
print("-"*65)
for i in d.keys():
    print(i,' : ',d[i])
```
