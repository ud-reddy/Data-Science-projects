```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("country_vaccinations.csv")
```

```python
data = df.values
```

```python
df
```

# Sorting Countries 

## 1) Country having least total vaccinations 

```python
countries = list(set(list(df['country'])))    
countries.sort() # sorting countries in the list in Ascending order
```

```python
d=[]
for i in countries:
    l=[]
    for j in data:
        if(j[0]==i):
            l.append([j[0],j[3]])
    pdf=pd.DataFrame(l,columns=['country','total'])
    d.append([i,pdf.sort_values(by='total',ascending=True).values[0][1]])
fdf=pd.DataFrame(d,columns=['country','total'])
```

```python
fdf = fdf.sort_values(by='total',ascending=True)
```

```python
c=0
for i in range(len(list(fdf['country']))):
    if(list(fdf['total'])[i] != 0):
        c=i
        break
print('-'*60)
print("The following countries have the least total vaccinations (zero): ")
print('-'*60)
j=1
for i in range(c):
    print(j,' : ',fdf.values[i][0])
    j+=1
print('-'*60)
```

## 2) Country having the most total vaccinations

```python
d=[]
for i in countries:
    l=[]
    for j in data:
        if(j[0]==i):
            l.append([j[0],j[3]])
    pdf=pd.DataFrame(l,columns=['country','total'])
    d.append([i,pdf.sort_values(by='total',ascending=False).values[0][1]])
adf=pd.DataFrame(d,columns=['country','total'])
print('-'*60)
print("The Country with the most Total Vaccinatios : ")
print('-'*60)
print(adf.sort_values(by='total',ascending=False).values[0][0])
print('-'*60)
```

## 3) Country with most daily vaccinations 

```python
l=[]
print("The country with the most number of daily Vaccinations : ")
print('-'*60)
for i in data:
    l.append([i[0],i[6]])
pdf=pd.DataFrame(l,columns=["country",'daily_vaccinations'])
print(pdf.sort_values(by='daily_vaccinations',ascending=False).values[0][0]," : ",pdf.sort_values(by='daily_vaccinations',ascending=False).values[0][1])
print('-'*60)
```

## 4) Country with least daily vaccinations 

```python
l=[]
print("The country with the most number of daily Vaccinations : ")
print('-'*60)
for i in data:
    l.append([i[0],i[6]])
pdf=pd.DataFrame(l,columns=["country",'daily_vaccinations'])
print(pdf.sort_values(by='daily_vaccinations',ascending=True).values[0][0]," : ",pdf.sort_values(by='daily_vaccinations',ascending=True).values[0][1])
print('-'*60)
```

## 5) Country wise No of people fully vaccinated 

```python
d1=[]
d=[]
for i in data:
    d.append([i[0],i[5]])
fdf=pd.DataFrame(d,columns=['country','full_vaccinations'])
for i in countries:
    l=[]
    for j in fdf.values:
        if(i==j[0]):
            l.append(j)
    tdf=pd.DataFrame(l,columns=['country','full_vaccinations'])
    d1.append([tdf.sort_values(by='full_vaccinations',ascending=False).values[0][0],tdf.sort_values(by='full_vaccinations',ascending=False).values[0][1]])
fullvaccinations=pd.DataFrame(d1,columns=['country','people_fully_vaccinated'])
fullvaccinations
            
```
